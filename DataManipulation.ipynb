```python
import pandas as pd

df = pd.read_csv('Details.csv')
df.drop(columns=['PaymentMode'], inplace=True)

print(df.head())
```

```python
df.describe()
```

```python
# Handling missing values
df
```

```python
#Filling missing values
#Updating column values
df = df.rename(columns={'Amount': 'Amount No'})

df['New Amount'] = df['Amount No'] 
df['New Amount'] = df['New Amount'].fillna(df['New Amount'].mean())
df['New Amount'] = df['New Amount'].astype(float)
df
```

```python
# Applying a function to a column
df['Quantity'] = df['Quantity'].apply(lambda x : x * 2)
df
```

```python
# Data aggregation and grouping

grouped_data = df.groupby('Category')['Profit'].mean()
print(grouped_data)
```
